```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

##solution 1, 2, 3 representing SOM binding weight-0.25,0.5,1--reactive sites bind with Ca2+
output1 = np.loadtxt('clay_fresh_high.txt', skiprows=1)
output2 = np.loadtxt('clay_fresh_low.txt', skiprows=1)
output3 = np.loadtxt('clay_fresh_ref.txt', skiprows=1)
obv=np.loadtxt('clay_fresh.csv',delimiter=",",skiprows=1)

plt.figure()

fig, ax=plt.subplots(3,3)
plt.subplot(331)
plt.plot(output1[:,0],'#fee0b6',output2[:,0],'#e08214',output3[:,0],'#7f3b08')
plt.plot(obv[:,0],obv[:,1],"o")
plt.title('pH')

plt.subplot(332)
plt.plot(output1[:,10]/1000,'#fee0b6',label="high")
plt.plot(output2[:,10]/1000,'#e08214',label="low")
plt.plot(output3[:,10]/1000,'#7f3b08',label="ref")
plt.legend(loc="upper left")
plt.title('SC (ds/m)')

plt.subplot(333)
plt.plot(output1[:,8]*1000,'#fee0b6',output2[:,8]*1000,'#e08214',output3[:,8]*1000,'#7f3b08')
plt.plot(obv[:,0],obv[:,8],"o")
plt.title('NH4(mM)')

plt.subplot(334)
plt.plot(output1[:,3]*1000,'#fee0b6',output2[:,3]*1000,'#e08214',output3[:,3]*1000,'#7f3b08')
plt.plot(obv[:,0],obv[:,4],"o")
plt.title('Ca(mM)')

plt.subplot(335)
plt.plot(output1[:,4]*1000,'#fee0b6',output2[:,4]*1000,'#e08214',output3[:,4]*1000,'#7f3b08')
plt.plot(obv[:,0],obv[:,5],"o")
plt.title('Mg(mM)')

plt.subplot(336)
plt.plot(output1[:,1]*1000,'#fee0b6',output2[:,1]*1000,'#e08214',output3[:,1]*1000,'#7f3b08')
plt.plot(obv[:,0],obv[:,2],"o")
plt.title('Na(mM)')

plt.subplot(337)
plt.plot(output1[:,7]*1000,'#fee0b6',output2[:,7]*1000,'#e08214',output3[:,7]*1000,'#7f3b08')
plt.plot(obv[:,0],obv[:,9],"o")
plt.title('SO4(mM)')

plt.subplot(338)
plt.plot(output1[:,6]*1000,'#fee0b6',output2[:,6]*1000,'#e08214',output3[:,6]*1000,'#7f3b08')
plt.plot(obv[:,0],obv[:,7],"o")
plt.title('Cl (mM)')

plt.subplot(339)
plt.plot(output1[:,5]*1000,'#fee0b6')
plt.plot(output2[:,5]*1000,'#e08214')
plt.plot(output3[:,5]*1000,'#7f3b08')
plt.plot(obv[:,0],obv[:,6],"o",label='Obv')
plt.title('PO4(mM)')
plt.suptitle("clay_baserun_salt")

fig.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

##solution 1, 2, 3 representing SOM binding weight-0.25,0.5,1--reactive sites bind with Ca2+
output1 = np.loadtxt('vanD_clay_fresh_a.txt', skiprows=1)
output3 = np.loadtxt('vanD_clay_salt_a.txt', skiprows=1)
obv1=np.loadtxt('clay_fresh.csv',delimiter=",",skiprows=1)
obv3=np.loadtxt('clay_salt.csv',delimiter=",",skiprows=1)
output1ex = np.loadtxt('vanD_clay_fresh_c.txt', skiprows=1)
output3ex = np.loadtxt('vanD_clay_salt_c.txt', skiprows=1)


plt.figure()

fig, ax=plt.subplots(3,3)
plt.subplot(331)
plt.plot(output1[:,0],'#fee0b6',output3[:,0],'#7f3b08', linestyle="dashed")
plt.plot(output1[:,0],'#fee0b6',output3[:,0],'#7f3b08')
plt.plot(obv1[:,0],marker="o",color='#fee0b6', linestyle="none")
plt.plot(obv3[:,0],color='#7f3b08',marker="o", linestyle="none")
plt.title('pH')

plt.subplot(332)
plt.plot(output1[:,10]/1000,'#fee0b6',output3[:,10]/1000,'#7f3b08', linestyle="dashed")
plt.plot(output1ex[:,10]/1000,'#fee0b6',output3ex[:,10]/1000,'#7f3b08')
plt.title('SC (ds/m)')

plt.subplot(333)
plt.plot(output1[:,8]*1000,'#fee0b6',output3[:,8]*1000,'#7f3b08', linestyle="dashed")
plt.plot(output1ex[:,8]*1000,'#fee0b6',output3ex[:,8]*1000,'#7f3b08')
plt.plot(obv1[:,7],marker="o",color='#fee0b6', linestyle="none")
plt.plot(obv3[:,7],color='#7f3b08',marker="o", linestyle="none")
plt.title('NH4(mM)')

plt.subplot(334)
plt.plot(output1[:,3]*1000,'#fee0b6',output3[:,3]*1000,'#7f3b08', linestyle="dashed")
plt.plot(output1ex[:,3]*1000,'#fee0b6',output3ex[:,3]*1000,'#7f3b08')
plt.plot(obv1[:,3],marker="o",color='#fee0b6', linestyle="none")
plt.plot(obv3[:,3],color='#7f3b08',marker="o", linestyle="none")
plt.title('Ca(mM)')

plt.subplot(335)
plt.plot(output1[:,4]*1000,'#fee0b6',output3[:,4]*1000,'#7f3b08', linestyle="dashed")
plt.plot(output1ex[:,4]*1000,'#fee0b6',output3ex[:,4]*1000,'#7f3b08')
plt.plot(obv1[:,4],marker="o",color='#fee0b6', linestyle="none")
plt.plot(obv3[:,4],color='#7f3b08',marker="o", linestyle="none")
plt.title('Mg(mM)')

plt.subplot(336)
plt.plot(output1[:,1]*1000,'#fee0b6',output3[:,1]*1000,'#7f3b08')
plt.plot(output1ex[:,1]*1000,'#fee0b6',output3ex[:,1]*1000,'#7f3b08', linestyle="dashed")
plt.plot(obv1[:,1],marker="o",color='#fee0b6', linestyle="none")
plt.plot(obv3[:,1],color='#7f3b08',marker="o", linestyle="none")
plt.title('Na(mM)')

plt.subplot(337)
plt.plot(output1[:,7]*1000,'#fee0b6',output3[:,7]*1000,'#7f3b08', linestyle="dashed")
plt.plot(output1ex[:,7]*1000,'#fee0b6',output3ex[:,7]*1000,'#7f3b08')
plt.plot(obv1[:,8],marker="o",color='#fee0b6', linestyle="none")
plt.plot(obv3[:,8],color='#7f3b08',marker="o", linestyle="none")
plt.title('SO4(mM)')

plt.subplot(338)
plt.plot(output1[:,6]*1000,'#fee0b6',output3[:,6]*1000,'#7f3b08', linestyle="dashed")
plt.plot(output1ex[:,6]*1000,'#fee0b6',output3ex[:,6]*1000,'#7f3b08')
plt.plot(obv1[:,6],marker="o",color='#fee0b6', linestyle="none")
plt.plot(obv3[:,6],color='#7f3b08',marker="o", linestyle="none")
plt.title('Cl (mM)')

plt.subplot(339)
plt.plot(output1[:,5]*1000,'#fee0b6', linestyle="dashed")
plt.plot(output3[:,5]*1000,'#7f3b08', linestyle="dashed")
plt.plot(output1ex[:,5]*1000,'#fee0b6',output3ex[:,5]*1000,'#7f3b08')
plt.plot(obv1[:,5],marker="o",color='#fee0b6', linestyle="none")
plt.plot(obv3[:,5],color='#7f3b08',marker="o", linestyle="none")
plt.title('PO4(mM)')
#plt.legend(loc="upper left")
plt.suptitle("Peat fresh (light)-salty (dark)")

fig.tight_layout()
plt.show()
```
