```python

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

##solution 1, 2, 3 representing SOM binding weight-0.25,0.5,1--reactive sites bind with Ca2+
output1 = np.loadtxt('vanD_peat_fresh_exch_a.txt', skiprows=1)
output2 = np.loadtxt('vanD_peat_fluc_exch_a.txt', skiprows=1)
output3 = np.loadtxt('vanD_peat_salt_exch_a.txt', skiprows=1)

plt.figure()

fig, ax=plt.subplots(3,3)
plt.subplot(331)
plt.plot(output1[:,0],'#fee0b6',output2[:,0],'#e08214',output3[:,0],'#7f3b08')
plt.title('pH')

plt.subplot(332)
plt.plot(output1[:,10]/1000,'#fee0b6',output2[:,10]/1000,'#e08214',output3[:,10]/1000,'#7f3b08')
plt.title('SC (ds/m)')

plt.subplot(333)
plt.plot(output1[:,8]*1000,'#fee0b6',output2[:,8]*1000,'#e08214',output3[:,8]*1000,'#7f3b08')
plt.title('NH4(mM)')

plt.subplot(334)
plt.plot(output1[:,3]*1000,'#fee0b6',output2[:,3]*1000,'#e08214',output3[:,3]*1000,'#7f3b08')
plt.title('Ca(mM)')

plt.subplot(335)
plt.plot(output1[:,4]*1000,'#fee0b6',output2[:,4]*1000,'#e08214',output3[:,4]*1000,'#7f3b08')
plt.title('Mg(mM)')

plt.subplot(336)
plt.plot(output1[:,1]*1000,'#fee0b6',output2[:,1]*1000,'#e08214',output3[:,1]*1000,'#7f3b08')
plt.title('Na(mM)')

plt.subplot(337)
plt.plot(output1[:,7]*1000,'#fee0b6',output2[:,7]*1000,'#e08214',output3[:,7]*1000,'#7f3b08')
plt.title('SO4(mM)')

plt.subplot(338)
plt.plot(output1[:,9]*1000,'#fee0b6',output2[:,9]*1000,'#e08214',output3[:,9]*1000,'#7f3b08')
plt.title('HCO3 (mM)')

plt.subplot(339)
plt.plot(output1[:,5]*1000,'#fee0b6',label='fresh')
plt.plot(output2[:,5]*1000,'#e08214',label='fluc')
plt.plot(output3[:,5]*1000,'#7f3b08',label='salt')
plt.title('PO4(mM)')
plt.legend(loc="upper left")
plt.suptitle("Peat with EXC=0")

fig.tight_layout()
plt.show()
```
